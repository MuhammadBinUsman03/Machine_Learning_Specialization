## **FURTHER ANALYSIS**
# **PREDICT GOOGLE PLAYSTORE RATINGS**
We have successfully done exploratory data analysis on above data. Now Our objective is to develop a predictive model that can accurately predict app ratings based on various features and attributes present in the Google Playstore dataset. By training this model, we aim to provide app developers and stakeholders with a tool to estimate the potential ratings of their apps and make informed decisions to improve user satisfaction and app performance.

Based on the heatmap results from above and the importance of category attribute for  the user; we decided to include all the  four attributes (`Rating Count`, `size`, `installs`,`Content rating`, and `category` ) in defining the app profile
whereas `ratings` will be choosen as the label for our model

We will be importing python library `sklearn` which provides simple and efficient tools for predictive data analysis

```python
from sklearn import metrics
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

Following above data cleaning activities , our columns `size`,`installs` are now ready to be fed into the model but we still need to one-hot encode `category`and `Content rating`.

```python
# Category features encoding
category_list = df['Category'].unique().tolist()
category_list = ['cat_' + word for word in category_list]
df = pd.concat([df, pd.get_dummies(df['Category'], prefix='cat')], axis=1)
```

```python
# Encode Content Rating features
le = preprocessing.LabelEncoder()
df['Content Rating'] = le.fit_transform(df['Content Rating'])
```

```python
# Split data into training and testing sets
features = ['Rating Count', 'Size', 'Installs', 'Content Rating',]
features.extend(category_list)
df.fillna({'Rating':df['Rating'].mean(),'Rating Count':df['Rating Count'].mean()},inplace=True) #Just for the model to work correctly otherwise ratings should be filled by other business techniques
X = df[features]
y = df['Rating']
```

```python
# Split the dataset into 75% train data and 25% test data.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

We can use KNN solve complicated problems. With the help of KNN we can do pattern recognition and data mining. KNN defines the similarity. From the given dataset KNN finds common groups between attributes

```python
# Look at the 8 closest neighbors
model = KNeighborsRegressor(n_neighbors=8)
df.isnull().sum().sort_values(ascending = False)
```

```python
# Find the mean accuracy of knn regression using X_test and y_test
model.fit(X_train, y_train)
```

```python
# Calculate the mean accuracy of the KNN model
accuracy = model.score(X_test,y_test)
'Accuracy: ' + str(np.round(accuracy*100, 2)) + '%'
```

Hence Now our model can predict ratings for unseen apps with 93% accuracy

```python
# Try different numbers of n_estimators - this will take a minute or so
n_neighbors = np.arange(1, 20, 1)
scores = []
for n in n_neighbors:
    model.set_params(n_neighbors=n)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))
plt.figure(figsize=(7, 5))
plt.title("Effect of Estimators")
plt.xlabel("Number of Neighbors K")
plt.ylabel("Score")
plt.plot(n_neighbors, scores)
```
